```python
rng = default_rng(340)
plt.figure()

ax1 = plt.subplot(2, 2, (1, 2))
for N in range(500,5001, 500):
    trials = 500
    pi_trial = np.zeros(trials)
    for i in range(trials):
        x = rng.random(N)
        y = rng.random(N)

        r = x**2 + y**2

        pi_trial[i] = np.sum(r < 1**2)/N*4

    mean_pi = np.mean(pi_trial)
    std_pi = np.std(pi_trial)

    ax1.plot(N*np.ones(trials), pi_trial, 'ko', alpha = 0.1)
plt.title(r'Value of $\pi=${:1.3f}$\pm${:1.3f}'.format(mean_pi, std_pi));
plt.xlabel('Trials');
plt.ylabel('Frequency');
ax2 = plt.subplot(2, 2, 3)
ax2.hist(pi_trial, 20, density=True)


x = np.linspace(3, 3.3)
pi_pdf = stats.norm.pdf(x, loc = mean_pi, scale = std_pi)
ax2.plot(x, pi_pdf)
plt.xlabel('Distribution');
plt.ylabel('Frequency');

ax3 = plt.subplot(2, 2, 4)
ax3.boxplot(pi_trial, vert=False);
```

$\pi$ is not a random variable. $\pi$ is the ratio of a circle's circumference to its diameter. $\pi$ can be solved using Monte Carlo methods. The Monte Carlo method, used random sampling to estimate $\pi$. Using Monte Carlo methods, random points along the x and y axis are generated between 0 and 1. The points input in equation x^2 + y^2, to calculate the radius. If the radius is greater than 1, the point is outside of the circle. If the radius is less than 1, the point is inside of the circle. The ratio of the number of points inside the circle to the total number of points is then multiplied by 4. This gives an estimate of $\pi$.
